```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/home/klay/ESTUDOS/PANDAS/app/app_controller/Banco de Ninjas.csv")
df["Salario"] = pd.to_numeric(df["Salario"],errors = "coerce")
media = df["Salario"].mean()
df = df.drop(6)
df = df.reset_index(drop = True)
df["Salario"] = df["Salario"].fillna(media)
print(df)
```

```python
for name in df["Nome"]:
    plt.plot(df["Nome"],df["Salario"],".",label=name)
```

```python
plt.hist(df["Rank"],bins = 10,color = 'r')
```

```python
plt.scatter(df["Rank"],df["Salario"])
```

```python
a = np.array(df["Rank"])
print(df.loc[df.query("Rank == 2").index]["Nome"].tolist())
v = np.unique_counts(a)
labels = []
plt.title("Ninjas Distribution")
plt.style.use('ggplot')
for item in v.values:
    labels.append('Rank {item}'.format(item=item))
plt.pie(v.counts,labels=labels,autopct='%.1f%%')
plt.show()
```

```python
ps = np.array(df["Rank"])
py = np.array(df["Salario"])
plt.hist2d(ps,py,bins=10,cmap='jet')
plt.xlabel('Rank')
plt.ylabel('Salario')
plt.xticks(np.arange(2,5,1))
plt.title("Ninjas Distribution")
plt.style.use('ggplot')
plt.title("Ninjas Distribution")
plt.show()
```

```python
plt.plot(df["Rank"],df["Salario"],marker= 'o')
plt.title("DistribuiÃ§Ã£o de SalÃ¡rios por Rank")
plt.xlabel("Rank")
plt.ylabel("Salario")
```

```python
 # X IS OUR PREDICTIVE VARIABLE
x = np.array(df["Rank"].values,int)
# Y IS OUR DEPENDENT VARIABLE
y = np.array(df["Salario"].values)
# BETAS ARE OUR LINEAR FUNCTION SLOPE AND INTERCEPT
betas = np.polyfit(x,y,1)
reta = np.poly1d(betas)

# WE SCATTER OUR DATA ALL ACROSS A CHART, THEN WE CROSS IT WITH OUR LINEAR REGRESSION FUNCTION STRAIGHT
plt.title("REGRESSION STRAIGHT")
plt.xlabel("Rank")
plt.ylabel("Salario esperado")
plt.xticks(np.arange(0,10,1))
plt.scatter(x,y)
plt.plot(x,reta(x))
plt.show()
```
